```python
school = pd.read_csv("schools_complete.csv")
student = pd.read_csv("students_complete.csv")
school = school.rename(columns={'name': 'school_name'})
student =student.rename(columns={'name': 'student_name',"school":"school_name"})
df = pd.merge(school,student, how = "left",on = ["school_name","school_name"])
df
```

```python
total_school = len((df.school_name).unique())
total_student = len(df.student_name)
total_budget = df.budget.sum()
Average_Math_Score = df.math_score.mean()
Average_Reading_Score = df.reading_score.mean()
#mathscoregreater than 50 is considered as pass
percentage_Passing_Math = len(df['math_score'].loc[df['math_score'] >= 50]) / len(df['math_score']) * 100
#reading score greater than 50 is considered as pass
percentage_Passing_Reading = len(df['reading_score'].loc[df['reading_score'] >= 50]) / len(df['reading_score']) * 100
#Overall Passing Rate (Average of the above two)
Overall_passing = (percentage_Passing_Math + percentage_Passing_Reading)/2
School_Summary = pd.DataFrame({"Total_school" : [total_school],
                            "total_student" : [total_student],
                            "total_budget": [total_budget],
                            "Average_Math_Score": [Average_Math_Score],
                            "Average_Reading_Score" : [Average_Reading_Score],
                            "percentage_Passing_Math" : [percentage_Passing_Math],
                            "percentage_Passing_Reading" :[percentage_Passing_Reading]                            })
School_Summary
```

# each school summary

Create an overview table that summarizes key metrics about each school, including:
School Name
School Type
Total Students
Total School Budget
Per Student Budget
Average Math Score
Average Reading Score
% Passing Math
% Passing Reading
Overall Passing Rate (Average of the above two)
Top Performing Schools (By Passing Rate)


```python

School_Name1 = school.set_index("school_name")['type']

Total_Students = df["school_name"].value_counts() 

Total_School_Budget = df.groupby(["school_name"]).mean()["budget"]

Per_Student_Budget = Total_School_Budget/Total_Students
Average_Math_Score =  df.groupby(["school_name"]).mean()["math_score"]
Average_Reading_Score = df.groupby(["school_name"]).mean()["reading_score"]

math_pass_score = df[(df["math_score"] >= 80)]
Passing_Math =  math_pass_score.groupby(["school_name"]).count()["student_name"]/Total_Students * 100
               
reading_pass_score = df[(df["reading_score"] >= 80)]
Passing_reading = reading_pass_score.groupby(["school_name"]).count()["student_name"]/Total_Students * 100

Overall_pass = df.loc[(df["math_score"] >= 80) & (df["reading_score"] >= 80)]
Overall  = Overall_pass.groupby(["school_name"]).count()["student_name"] / Total_Students * 100


each_school_summary = pd.DataFrame({"school_type" : School_Name1,
                                    "Total_students" : Total_Students,
                                    "Total_School_Budget" : Total_School_Budget,
                                    "Per_Student_Budget" : Per_Student_Budget,
                                    "Average_Math_Score" : Average_Math_Score,
                                    "Average_Reading_Score" : Average_Reading_Score,
                                    "Passing_Math" : Passing_Math,
                                    "Passing_reading" : Passing_reading,
                                    "Overall" : Overall})
each_school_summary
```
